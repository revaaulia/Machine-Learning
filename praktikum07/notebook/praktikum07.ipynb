# 8. Menentukan Fitur (X) dan Target (Y)

```python
# Menentukan fitur
X = df[['b12', 'b11', 'b9', 'b8a', 'b8', 'b7', 'b6', 'b5', 'b4', 'b3', 'b2', 'b1', 'Sigma_VV',
        'Sigma_VH', 'plia', 'lia', 'iafe', 'gamma0_vv', 'gamma0_vh', 'beta0_vv', 'beta0_vh']]

# Kolom target (label)
y = df['N']
```

**Penjelasan:**

Kode di atas digunakan untuk memisahkan data menjadi fitur dan target sebelum membuat model prediksi. Variabel **"X"** berisi fitur-fitur atau variabel input yang terdiri dari berbagai kolom seperti band spektral (b1-b12), data SAR (Sigma_VV, Sigma_VH), dan koefisien backscatter (gamma0, beta0). Sedangkan variabel **"y"** berisi kolom N, merupakan label target yang akan diprediksi model. Langkah ini dilakukan dalam persiapan data sebelum melatih model prediktif.

# 9. Split Data (Train) dan (Test)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=7
    )
```

**Penjelasan:**

Fungsi `train_test_split` digunakan untuk membagi data X dan y menjadi dua bagian yaitu X_train dan y_train (80% untuk melatih model), serta X_test dan y_test (20% untuk menguji model). Pembagian ini penting dilakukan agar model dapat dievaluasi menggunakan data yang belum pernah dilihat sebelumnya.

# 10. Jumlah Data Train dan Test

```python
# Menampilkan jumlah data train dan test
print(f"Jumlah data train: {len(X_train)}")
print(f"Jumlah data test: {len(X_test)}")
```

**Penjelasan:**

Kode di atas berfungsi untuk melihat jumlah banyaknya data yang sudah dibagi untuk (train) dan (test). Hasilnya menunjukkan bahwa model akan dilatih menggunakan 475 data, kemudian diuji menggunakan 119 data yang terpisah.

# 11. Membuat dan Melatih Model OLS

```python
# Membuat dan melatih (fit) model OLS
model_ols = sm.OLS(y_train, X_train).fit()
```

```python
# Menambahkan kolom konstanta
X = sm.add_constant(X)
```

**Penjelasan:**

Kode di atas digunakan untuk membuat dan melatih model regresi linear menggunakan metode OLS (Ordinary Least Squares) dari library `statsmodels`. Di dalamnya, dimasukkan data **y_train** (sebagai target yang akan diprediksi) dan **X_train** (sebagau fitur atau variabel prediktor) yang telah disiapkan sebelumnya. Perintah `.fit()` akan menjalankan perhitungan statistik untuk mencari garis regresi terbaik yang paling sesuai dengan data pelatihan. Hasilnya disimpan dalam variabel **model_ols**, yang selanjutnya dapat digunakan untuk melihat ringkasan hasil analisis dan juga melakukan prediksi terhadap data baru.

Kemudian, dilakukan penambahan kolom "konstanta" yang diperlukan dalam model regresi statistik karena kolom ini akan menambahkan satu kolom baru berisi angka 1 ke dalam data fitur.


# 12. Prediksi Menggunakan Data Uji

```python
# Melakukan prediksi pada data uji
y_pred = model_ols.predict(X_test)
```

**Penjelasan:**

Kode `y_pred = model_ols.predict(X_test)` digunakan untuk melakukan prediksi menggunakan model regresi linear OLS yang telah dilatih sebelumnya. Dalam hal ini, **`model_ols`** merupakan model yang sudah mempelajari hubungan antara variabel input (**X_train**) dan target (**y_train**) saat proses pelatihan. Kemudian, fungsi `.predict(X_test)` digunakan untuk memasukkan data uji **(X_test)** agar model menghasilkan nilai prediksi **(y_pred)** berdasarkan pola yang telah dipelajari. Hasil prediksi berupa nilai prediksi target yang bisa dibandingkan dengan nilai sebenarnya (**y_test**) untuk mengukur seberapa baik model dalam memprediksi data baru.

# 13. Menghitung Metrik Evaluasi

```python
# Menghitung metrik evaluasi
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Cetak hasil
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (RÂ²): {r2:.4f}")
print("-" * 40)
```

**Penjelasan:**

Kode di atas dijalankan untuk mengevaluasi tingkat akurasi model dengan membandingkan hasil prediksi (**y_pred**) terhadap nilai sebenarnya (**y_test**) pada data uji. Terdapat tiga metrik evaluasi yang digunakan, yaitu **MSE (Mean Squared Error)** yang mengukur rata-rata selisih kuadrat, **RMSE (Root Mean Squared Error)** yang merupakan akar dari MSE untuk menunjukkan rata-rata besar kesalahan dalam data, serta **R-squared (RÂ²)** yang menunjukkan seberapa besar variasi nilai target dapat dijelaskan oleh model.

Dari hasil evaluasi, nilai **RÂ² sebesar 0.5868**, berarti sekitar **58,68%** variasi pada data uji dapat dijelaskan oleh model, nilai ini menunjukkan performa yang cukup baik, namun tidak sempurna sehingga masih bisa ditingkatkan. Nilai **RMSE sebesar 0.2630**, menunjukkan rata-rata jarak prediksi dari nilai sebenarnya sekitar 0.26 satuan, sehingga model cukup akurat. Sedangkan, MSE (0.0692) berarti rata-rata kesalahan kuadrat antara nilai prediksi dan nilai aktual cukup kecil.

Kemudian, pada kolom **P>|t|** menunjukkan variabel mana yang paliing berpengaruh signifikan, jika angkanya di bawah < 0.05, maka variabel tersebut benar-benar berpengaruh terhadap hasil prediksi model.
